```python
'''
    1. Develop a Python Tkinter application that takes two numbers as input, adds them, and displays the result.
'''


import tkinter as tk
from tkinter import messagebox

# Function to add two numbers
def add_numbers():
    try:
        num1 = float(entry1.get())
        num2 = float(entry2.get())
        result = num1 + num2
        result_label.config(text=f"Result: {result}")
    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter valid numbers.")

# Create main window
root = tk.Tk()
root.title("Simple Adder")
root.geometry("300x200")

# Labels and entries
tk.Label(root, text="Enter first number:").pack(pady=5)
entry1 = tk.Entry(root)
entry1.pack()

tk.Label(root, text="Enter second number:").pack(pady=5)
entry2 = tk.Entry(root)
entry2.pack()

# Button to add numbers
tk.Button(root, text="Add", command=add_numbers).pack(pady=10)

# Label to show result
result_label = tk.Label(root, text="Result: ")
result_label.pack(pady=10)

# Run the application
root.mainloop()
```

```python
'''
    4.Design and implement a Python GUI application using Tkinter that enables new users to register
      by entering their personal details.
      The GUI should include:
      • Title: “Examination Registration Form”
      • Labels and Entry Fields for the following details:
      • Full Name
      • Roll Number
      • Course Name
      • Subject
      • Semester
      • Email Address
      • Buttons:
      • Register (to submit the form)
      • Clear (to reset all fields)
      When the Register button is clicked:

      • Validate that no field is left empty.
      • If all fields are filled:
      • Display a message box: “Registration Successful!”
      • If any field is empty:
      • Display an error message: “All fields are required!”
      • When the Clear button is clicked:
      • All entry fields should be cleared/reset.
'''

import tkinter as tk
from tkinter import messagebox

# Function to handle registration
def register():
    # Get all field values
    name = entry_name.get()
    roll = entry_roll.get()
    course = entry_course.get()
    subject = entry_subject.get()
    semester = entry_semester.get()
    email = entry_email.get()

    # Check if any field is empty
    if not (name and roll and course and subject and semester and email):
        messagebox.showerror("Error", "All fields are required!")
    else:
        messagebox.showinfo("Success", "Registration Successful!")

# Function to clear all fields
def clear_fields():
    entry_name.delete(0, tk.END)
    entry_roll.delete(0, tk.END)
    entry_course.delete(0, tk.END)
    entry_subject.delete(0, tk.END)
    entry_semester.delete(0, tk.END)
    entry_email.delete(0, tk.END)

# Create main window
root = tk.Tk()
root.title("Examination Registration Form")
root.geometry("400x400")

# Title Label
title_label = tk.Label(root, text="Examination Registration Form", font=("Helvetica", 14, "bold"))
title_label.pack(pady=10)

# Frame for form fields
form_frame = tk.Frame(root)
form_frame.pack(pady=10)

# Labels and Entry Fields
tk.Label(form_frame, text="Full Name:").grid(row=0, column=0, sticky="w", pady=5)
entry_name = tk.Entry(form_frame, width=30)
entry_name.grid(row=0, column=1)

tk.Label(form_frame, text="Roll Number:").grid(row=1, column=0, sticky="w", pady=5)
entry_roll = tk.Entry(form_frame, width=30)
entry_roll.grid(row=1, column=1)

tk.Label(form_frame, text="Course Name:").grid(row=2, column=0, sticky="w", pady=5)
entry_course = tk.Entry(form_frame, width=30)
entry_course.grid(row=2, column=1)

tk.Label(form_frame, text="Subject:").grid(row=3, column=0, sticky="w", pady=5)
entry_subject = tk.Entry(form_frame, width=30)
entry_subject.grid(row=3, column=1)

tk.Label(form_frame, text="Semester:").grid(row=4, column=0, sticky="w", pady=5)
entry_semester = tk.Entry(form_frame, width=30)
entry_semester.grid(row=4, column=1)

tk.Label(form_frame, text="Email Address:").grid(row=5, column=0, sticky="w", pady=5)
entry_email = tk.Entry(form_frame, width=30)
entry_email.grid(row=5, column=1)

# Buttons Frame
button_frame = tk.Frame(root)
button_frame.pack(pady=20)

# Buttons
tk.Button(button_frame, text="Register", command=register, width=12, bg="lightgreen").grid(row=0, column=0, padx=10)
tk.Button(button_frame, text="Clear", command=clear_fields, width=12, bg="lightcoral").grid(row=0, column=1, padx=10)

# Run the GUI
root.mainloop()
```
